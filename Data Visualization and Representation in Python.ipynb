# Data Visualization and Representation in Python
The Anscombe Quartet and the importance of visualizing data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize=(7,4))
fuel_unit = pd.DataFrame({'unit':['BBL','GAL','GRAMSU','KGU','MCF','MMBTI','MWDTH','MWTH','TON'],
                         'COUNT':[7998,84,464,110,11354,180,95,100,8958]})
sns.barplot(data = fuel_unit, x = 'unit', y='COUNT')
plt.xlabel('fuel Unit')

```

```python
# Using logarithm to plot the barchart for the y-axis , beacue of the extreme rabge of the values formthe fuel unit
g =sns.barplot(data = fuel_unit, x = 'unit', y= 'COUNT')
g.set_yscale('log')
g.set_ylim(1,12000)
plt.xlabel('fuel unit')
```

```python
fuel_df=pd.read_csv('https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv')
fuel_df
```

- Advanced plotting: Kernel Density Estimate plots, box plots and violin plots

```python
# Select a sample of the dataset
sample_df = fuel_df.sample(n=50, random_state=4)
sns.regplot(x=sample_df["utility_id_ferc1"], y=sample_df["fuel_cost_per_mmbtu"],
fit_reg=False)
```

```python
# Box plot
sns.boxplot(x="fuel_type_code_pudl", y="utility_id_ferc1",
palette=["m", "g"], data=fuel_df)
```

```python
# KDE plot
sns.kdeplot(sample_df['fuel_cost_per_unit_burned'], shade=True, color="b")
```

```python
# Plotting a violin plot
sns.violinplot(x="fuel_qty_burned", y="fuel_cost_per_unit_delivered", data=fuel_df)
plt. show()


```
